```python
import pandas as pd

data = pd.read_csv(r'C:\Users\Hp  User\Documents\Delf√≠n\Datos\Municipios2015.csv', encoding = 'utf8')
```

```python
Jalisco = data[data.Entidad == 'Jalisco']
```

```python
df_15 = Jalisco.loc[Jalisco['Municipio'].isin(['Guadalajara','Zapopan', "San Pedro Tlaquepaque"])]
```

```python
delitos = df_15.set_index(['Tipo de delito'])

delitos.drop(delitos.iloc[:, 0:8], inplace = True, axis = 1) 
```

```python
dfd_15 = delitos.groupby(level=-1, axis=0).sum()
```

```python
import matplotlib.pyplot as plt
import pylab as plot
import numpy as np
import calendar 

deli_15 = dfd_15.transpose()

deli_15.plot()
plt.xticks(np.arange(12), dfd_15, rotation=30)
plt.xlabel('Meses del a√±o')
plt.ylabel('N√∫mero de delitos')

leg = plt.legend(bbox_to_anchor=(1.04,1), loc="upper left", fontsize = 'small')
```
